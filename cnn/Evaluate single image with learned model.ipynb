```python
from keras.models import model_from_json
import numpy as np
from keras.preprocessing import image

model = model_from_json(open('F:/Google Drive/User_Backup/model/cnn_model4.json').read())
model.load_weights('F:/Google Drive/User_Backup/model/cnn_model_weights4.hdf5')

#Specify image that you would like to evaluate with the trained model.
#filename = 'F:/Google Drive/User_Backup/test_data/20_1.png'
filename = 'C:/Users/Hiroji/cutout/112.jpg'
img = image.load_img(filename, grayscale = True, color_mode ='grayscale', target_size=(25, 44))

x = image.img_to_array(img)
x /= 255
x = np.expand_dims(x, axis=0)
y_proba = model.predict(x)
result = y_proba.tolist()

#Evaluation result
print(model.predict_classes(x))
#print(result)
for i in range(100):
    print(i, result[0][int(i)])
```
